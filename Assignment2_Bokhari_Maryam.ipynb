<b> Importing Neccesary Libraries </b>


```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .master("local[*]") \
    .appName("Learning_Spark") \
    .getOrCreate()
sc= spark.sparkContext

from xml.dom import minidom
mydoc = minidom.parse('Users.xml')
User= mydoc.getElementsByTagName('row')
```

<b> Question 1 </b>From the Users.xml file, find all users which are from Georgia and output to screen their
DisplayName only.

```python
Loc_Display=[]
for i in range(len(User)):
    if('Location' in User[i].attributes.keys() and 'DisplayName' in User[i].attributes.keys() ):
        Loc_Display.append((User[i].attributes['Location'].value, User[i].attributes['DisplayName'].value)) 
data=sc.parallelize( Loc_Display )
rdd=data.filter(lambda x: "Georgia" in x[0])
for x in rdd.collect():
    print x[1]
```

<b>Question 2</b>  Using the Users.xml file, provide the count for all users which joined (CreationDate) in 2017. Output this to the screen

```python
date=[] 
for i in range(len(User)):
    if('CreationDate' in User[i].attributes.keys()):
        date.append((User[i].attributes['CreationDate'].value)) 
data=sc.parallelize( date )
#data.collect()
data.filter(lambda x: "2017" in x).count()
```

<b>Question 3 </b> Using the PostHistory file, count the number of Posts that feature the words “Spark” and
“Scala”. Output this to the screen. (20 points)


```python
PostHistory=[]
PostUsers=[]
mydoc = minidom.parse('PostHistory.xml')
History= mydoc.getElementsByTagName('row')
for i in range(len(History)):
    if('Text' in History[i].attributes.keys() and 'PostId' in History[i].attributes.keys() ):
        PostHistory.append(( History[i].attributes['Text'].value))
    if('Id' in History[i].attributes.keys()):
        PostUsers.append(( History[i].attributes['Id'].value))
```

```python
words=["Spark","Scala"]
a=sc.parallelize(PostHistory).filter(lambda x: words[0] in x).filter(lambda x: words[1] in x).count()
print(a)
```
